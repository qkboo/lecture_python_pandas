# 정규표현식 실습

html 파일의 태그 크기, 텍스트 크기, 전체 문자 길이 출력

```html
<doctype html>
<title>
<h1>
</h1>

<scipt></script>
```

````python
with open('news_naver.html', encoding='cp949') as f:
    html = f.read()
````

````python
len(html), html[:100]
````

````python
import re
````

#### `<body></body>` 의 내용만 잘라내기

````python
body = re.search('<body.*/body>', html, re.I|re.S) # re.S == re.DOTALL
body = body.group()
body
````

````python
len(body)
````

#### `<script>...</script>` 삭제

 - `<script>...</script>` --> ''

````python
body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I|re.S)
body
````

````python
len(body)
````

#### 태그, 주석...

```
<tag>
<tag>text</tag>
<tag attribte...></tag>
```

````python
text = re.sub('<.+?>', '', body, 0, re.I|re.S)
text[400:3000]
````

````python
text.split()
````

````python
nospace = re.sub('&nbsp;| |\t|\r|\n', '', text)
nospace[:300]
````

````python
print (f'html = {len(html)}, text = {len(text)}, word = {len(text.split())} nospace = {len(nospace)}')
````
